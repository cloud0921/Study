```python
import matplotlib.pyplot as plt
plt.title('legend')
plt.plot([10,20,30,40], label='asc') # 증가를 의미하는 asc
plt.plot([40,30,20,10], label='dsc') # 감소를 의미하는 dsc
plt.legend()
plt.show()
```
